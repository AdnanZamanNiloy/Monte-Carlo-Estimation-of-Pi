```python
import random
import matplotlib.pyplot as plt

# Input number of points from user
num_points = 10000

circle_points = 0
points_inside = []
points_outside = []

for _ in range(num_points):
    # Generate random numbers between -1 and 1
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)

    # Check if inside circle
    if x*x + y*y <= 1:
        circle_points += 1
        points_inside.append((x, y))
    else:
        points_outside.append((x, y))

# Pi estimation
pi_estimate = 4.0 * (circle_points / num_points)

print("Total Points:", num_points)
print("Inside Circle:", circle_points)
print("Estimated Pi:", pi_estimate)

# Plotting
xi = [p[0] for p in points_inside]
yi = [p[1] for p in points_inside]
xo = [p[0] for p in points_outside]
yo = [p[1] for p in points_outside]

plt.figure(figsize=(6, 6))
plt.scatter(xo, yo, s=1)
plt.scatter(xi, yi, s=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Monte Carlo Estimation of Pi")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
